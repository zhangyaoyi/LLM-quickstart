## Transformer Postion Encoding

```python
import torch
import math
import matplotlib.pyplot as plt

def get_positional_encoding(max_seq_len, embedding_dim):
    position = torch.arange(max_seq_len).unsqueeze(1)
    div_term = torch.exp(torch.arange(0, embedding_dim, 2) * (-math.log(10000.0) / embedding_dim))
    pos_encoding = torch.zeros(max_seq_len, embedding_dim)
    pos_encoding[:, 0::2] = torch.sin(position * div_term)
    pos_encoding[:, 1::2] = torch.cos(position * div_term)
    return pos_encoding

# 设置参数
max_seq_len = 10  # 最大序列长度
embedding_dim = 50  # 嵌入维度

# 生成位置编码
positional_encoding = get_positional_encoding(max_seq_len, embedding_dim)

# 打印位置编码的形状
print("Position Encoding Shape:", positional_encoding.shape)

print(positional_encoding)

# 可视化位置编码
plt.figure(figsize=(12, 8))
plt.imshow(positional_encoding, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Positional Encoding (718 dimensions)")
plt.xlabel("Embedding Dimension")
plt.ylabel("Sequence Position")
plt.tight_layout()
plt.show()
```

```python
import torch
import math

embedding_dim = 8  # 使用一个小的维度作为例子

# 步骤 1: 创建等差数列
step1 = torch.arange(0, embedding_dim, 2)
print("Step 1 - Arange:")
print(step1)

# 步骤 2: 计算常数
constant = -math.log(10000.0) / embedding_dim
print(f"\nStep 2 - Constant: {constant}")

# 步骤 3: 乘以常数
step3 = step1 * constant
print("\nStep 3 - Multiplication result:")
print(step3)

# 步骤 4: 取指数
div_term = torch.exp(step3)
print("\nStep 4 - Final div_term:")
print(div_term)

# 额外：检查 div_term 的形状
print("\nShape of div_term:", div_term.shape)
```
