```python
from transformers import pipeline

model_path = "/home/overman/Workspace/LLM-quickstart/transformers/models/bert-base-cased-finetune-yelp/checkpoint-40000"

#classifier = pipeline("text-classification", model="overman123/bert-base-cased-finetune-yelp", tokenizer="bert-base-cased", device="cuda")
classifier = pipeline("text-classification", model=model_path, tokenizer="bert-base-cased", device="cuda")

# 定义一个函数来分类文本
def classify_text(text):
    result = classifier(text)[0]
    return result['label'], result['score']

# 测试模型
test_texts = [
    "This restaurant is absolutely fantastic!",
    "The food was terrible and the service was even worse.",
    "I had a decent experience, nothing special but not bad either.",
    "The weather is so good today and I like it so much."
]

for text in test_texts:
    label, score = classify_text(text)
    print(f"文本: {text}")
    print(f"预测标签: {label}")
    print(f"置信度: {score:.4f}")
    print("---")
```
